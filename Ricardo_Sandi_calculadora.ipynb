```
def calcular(cantidad):
    resultado=0

    if cantidad == 2:
      
      x = input("Ingresa el primer valor: ")
      x = float(x)
      operacion = input("Ingresa la operacion a realizar (+, -, *, /)")
      y = input("Ingresa el segundo valor: ")
      y = float(y)

      resultado = float(resultado)

      if x % x == 1 and y % y ==1:
        x = int(x)
        y = int(y)
        resultado = int(resultado)
      
      if operacion == '+':
        resultado = x +  y
      elif operacion == '-':
        resultado = x -  y 
      elif operacion == '*':
        resultado = x *  y 
      else:
        resultado = x / y 

    elif cantidad == 3:

      x = input("Ingresa el primer valor: ")
      x = float(x)

      op1 = input("Ingresa la primera operacion a realizar (+, -, *, /)")

      y = input("Ingresa el segundo valor: ")
      y = float(y)

      op2 = input("Ingresa la segunda operacion a realizar (+, -, *, /)")

      z = input("Ingresa el tercer valor: ")
      z = float(z)

      resultado = float(resultado)

      if x % x == 1 and y % y ==1:
        x = int(x)
        y = int(y)
        resultado = int(resultado)

      if op1 == '+' and op2 == '+':
        resultado = x + y + z
      elif op1 == '+' and op2 == '-':
        resultado = x +  y - z
      elif op1 == '+' and op2 == '*':
        resultado = x + y * z
      elif op1 == '+' and op2 == '/':
        resultado = x + y / z

      elif op1 == '-' and op2 == '-':
        resultado = x -  y - z
      elif op1 == '-' and op2 == '+':
        resultado = x -  y + z
      elif op1 == '-' and op2 == '*':
        resultado = (x - y) * z
      elif op1 == '-' and op2 == '/':
        resultado = (x - y) / z

      elif op1 == '*' and op2 == '*':
        resultado = x * y * z
      elif op1 == '*' and op2 == '+':
        resultado = (x * y) + z
      elif op1 == '*' and op2 == '-':
        resultado = (x * y) - z
      elif op1 == '*' and op2 == '/':
        resultado = (x * y) / z

      elif op1 == '/' and op2 == '/':
        resultado = (x / y) / z
      elif op1 == '/' and op2 == '+':
        resultado = (x / y) + z
      elif op1 == '/' and op2 == '-':
        resultado = (x / y) - z
      elif op1 == '/' and op2 == '*':
        resultado = (x / y) * z


    return resultado
```

```
print("Esta calculadora realiza las operaciones una a una. Si deseas calcular aoperaciones \n" +
       "aritméticas con tres números se dará prioridad a la operación entre los dos primeros numeros elegidos y la primera operación. \n" +
        "A ese resultado se le sumará, restará, multiplicará o dividirá, según sea el caso, el tercer número ingresado \n" +
        "Por ejemplo: Primer valor: 4 \n" +
        "             Primera operación  : * \n" +
            "            Segundo valor: 2 \n" +
           "             Seunda operación  : + \n" +
            "            Tercer valor: 3 \n" +
           "Resultado: (4 * 2) + 3 = 11 \n"
           )

print("Cuantos valores desea incluir en el cálculo (2 o 3 )")
cant = input()
cantidad = int(cant)

print("Resultado: ", calcular(cantidad))
```
